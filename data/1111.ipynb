```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('正弦函数')
plt.grid(True)
plt.show()
```
