```python
import sys
import subprocess

print(f"当前Python解释器: {sys.executable}")
print(f"当前工作目录: {sys.path[0]}")

# 使用sys.executable来确保使用正确的Python
def safe_install(package):
    """安全安装包，避免路径空格问题"""
    cmd = [sys.executable, "-m", "pip", "install", package]
    print(f"执行: {' '.join(cmd)}")
    
    try:
        result = subprocess.run(cmd, capture_output=True, text=True)
        if result.returncode == 0:
            print(f"✅ {package} 安装成功")
            return True
        else:
            print(f"❌ {package} 安装失败")
            print(f"错误: {result.stderr[:200]}")
            return False
    except Exception as e:
        print(f"安装异常: {e}")
        return False

# 安装需要的包
packages = ['lxml', 'beautifulsoup4', 'html5lib']
for package in packages:
    safe_install(package)
```

```python
from lxml import objectify

xml = objectify.parse('examples/mta_perf/Performance_MNR.xml')
# # objectify.parse() 读取 XML 文件并返回一个对象树
root = xml.getroot()  
# getroot() 方法返回 XML 树的根元素
# 根据你的输出，这个根元素就是 <INDICATOR>
print(f"根元素是: {root.tag}") 
# # root.tag 获取根元素的标签名
```

```python
# 因为root就是INDICATOR，所以不需要循环
skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ', 'DESIRED_CHANGE', 'DECIMAL_PLACES']

el_data = {}
for child in root.getchildren():  # 直接遍历根元素的子元素
    if child.tag in skip_fields:
        continue
    # 处理值
    if child.text is None:
        value = ''
    else:
        value = child.text.strip()
    el_data[child.tag] = value

print("提取的数据:")
for key, value in el_data.items():
    print(f"{key}: {value}")

# 如果要放入列表（保持与其他代码兼容）
data = [el_data]
print(data)
```
